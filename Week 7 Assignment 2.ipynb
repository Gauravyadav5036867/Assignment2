```python
#Question 1

import requests
from bs4 import BeautifulSoup

def get_first_five_video_urls(channel_url):
    response = requests.get(channel_url)
    
    if response.status_code != 200:
        print(f"Failed to retrieve the page: {response.status_code}")
        return []
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Finding all video links
    video_links = soup.find_all('a', href=True)
    
    # Extract the URLs for the first five videos
    video_urls = []
    for link in video_links:
        href = link['href']
        if '/watch' in href and len(video_urls) < 5:
            video_urls.append(f"https://www.youtube.com{href}")
    
    return video_urls

channel_url = 'https://www.youtube.com/@PW-Foundation/videos'
video_urls = get_first_five_video_urls(channel_url)

for url in video_urls:
    print(url)
```

```python
#Question 2


from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

def get_first_five_thumbnail_urls(channel_url):
    # Set up Chrome options
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Ensure GUI is off
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    
    # Set path to chromedriver as per your configuration
    webdriver_service = Service('/path/to/chromedriver')  # Replace with your path to chromedriver
    
    # Choose Chrome Browser
    driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)
    
    # Open the YouTube channel page
    driver.get(channel_url)
    
    # Wait for the page to load
    time.sleep(5)  # You might need to adjust the sleep time depending on your internet speed
    
    # Extract video thumbnail URLs
    thumbnails = driver.find_elements(By.XPATH, '//*[@id="thumbnail"]')
    
    thumbnail_urls = []
    for thumbnail in thumbnails:
        if len(thumbnail_urls) < 5:
            img = thumbnail.find_element(By.TAG_NAME, 'img')
            thumbnail_url = img.get_attribute('src')
            if thumbnail_url:
                thumbnail_urls.append(thumbnail_url)
        else:
            break
    
    # Close the driver
    driver.quit()
    
    return thumbnail_urls

channel_url = 'https://www.youtube.com/@PW-Foundation/videos'
thumbnail_urls = get_first_five_thumbnail_urls(channel_url)

for url in thumbnail_urls:
    print(url)
```

```python
#Question 3

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

def get_first_five_video_titles(channel_url):
    # Set up Chrome options
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Ensure GUI is off
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    
    # Set path to chromedriver as per your configuration
    webdriver_service = Service('/path/to/chromedriver')  # Replace with your path to chromedriver
    
    # Choose Chrome Browser
    driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)
    
    # Open the YouTube channel page
    driver.get(channel_url)
    
    # Wait for the page to load
    time.sleep(5)  # You might need to adjust the sleep time depending on your internet speed
    
    # Extract video titles
    video_titles = []
    titles = driver.find_elements(By.XPATH, '//*[@id="video-title"]')
    
    for title in titles:
        if len(video_titles) < 5:
            video_titles.append(title.get_attribute('title'))
        else:
            break
    
    # Close the driver
    driver.quit()
    
    return video_titles

channel_url = 'https://www.youtube.com/@PW-Foundation/videos'
video_titles = get_first_five_video_titles(channel_url)

for title in video_titles:
    print(title)
```
